## Problem Statement:
We've to predict weather patient suffering from cancer or not.On given data We'll going to test differnt algorithm.

## Import Libraries

```python
# Import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import dataset

```python
data = pd.read_csv("data.csv")
data
```

## Data Exploration

```python
data.shape
```

```python
data.describe().T
```

```python
data.info()
```

```python
data.drop(columns=["id","Unnamed: 32"],inplace=True)
```

```python
data.columns
```

## Converting diagnosis column into integer

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["diagnosis"] = np.array(le.fit_transform(data["diagnosis"]))
data["diagnosis"]
```

## Splitting Dataset in X and y
* x = Features/Input
* Y = Lables/ Output

```python
x = data.drop(columns=["diagnosis"])
x
```

```python
y = data["diagnosis"]
y
```

## Splitting Data into Training and Test

We are giving 80 % data to training, and 20 % to test data.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Training Model

```python
#!pip install xgboost
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

## Prediction

```python
y_preds = model.predict(x_test)
y_preds
```

## Evaluting model

```python
## Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_preds)
print(cm)
```

```python
## Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_preds))
```

# Conclusion

Using **XGBOOST ALGORITITHM**, We got 97 % percent accuracy with is better as compare to other model.
