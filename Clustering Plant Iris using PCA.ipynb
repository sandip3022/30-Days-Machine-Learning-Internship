# Problem Statement:

In this notebook, I am going to perform **Plant Iris clustering with PCA**

# Importing Libraries: 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

# Dataset Segregation:

```python
x = dataset.data
y = dataset.target
```

```python
print(x)
```

```python
names = dataset.target_names
names
```

# Fitting the PCA clustering to dataset with n = 2

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)
y_means = model.fit(x).transform(x)
```

```python
y_means
```

# Ploting Cluster

```python
plt.figure()
colors = ["red","green","yellow"]

for color , i , target_names in zip(colors,[0,1,2],names):
        plt.scatter(y_means[y==i,0],y_means[y==i,1],color=color,lw=2,label=target_names)
plt.title("IRIS Clustering")
plt.legend()
plt.show()
```
